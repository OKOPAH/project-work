```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
dataset = pd.read_csv('red_wine.csv')
```

```python
dataset.head(10)
```

```python
nan_df = dataset[dataset.isna().any(axis=1)]
nan_df.head(10)
```

```python
dataset = dataset.dropna(how='any')
dataset.head(10)
```

```python
dataset.tail(20)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
quality_means = dataset.groupby('quality').mean()
print(quality_means)
```

```python
plt.figure(figsize=(12, 5))
plt.hist(dataset['alcohol'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Alcohol Content')
plt.xlabel('Alcohol Content (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot([dataset[dataset['pH'] == i]['alcohol'] for i in sorted(dataset['pH'].unique())],
            labels=sorted(dataset['pH'].unique()), showfliers=False, patch_artist=True, boxprops=dict(facecolor='blue', color='black'))

plt.title('Box Plot of Alcohol Content by pH')
plt.xlabel('Quality')
plt.ylabel('pH')
plt.show()
```
